# 入门理解

### 概念

- ** 特点 **

> 适用于非结构化数据存储的数据库

> 基于列的模式

> 适合处理的高扩展，低写入/查询延时

- ** 数据模型 **

> 数据模型类似于Google的BigTable.Bigtable是一个疏松的分布式的持久的多维排序的map,这个map被行键,列键,和时间戳索引.每一个值都是连续的byte数组


- ** hadoop 缺陷:　Hadoop 无法满足大规模数据实时处理 **

> hdfs适合存储非结构化数据

> Hadoop MapReduce 编程框架高延迟数据处理机制

> 


# hbase结构

- ** (Table, RowKey, Family, Column, Timestamp) -> Value **

- ** 列簇 **

> 在表创建时定义好, 并且不能修改太频繁, 数量也不能太多(只限于几十) 

> 所有列和行的信息都会通过列簇在表中定义

- ** 列 **

> 引用列的格式family:qualifier, qualifier是任意的字节数组, 列的数量没有限制: 一个列簇可以有数百万个列, 列值没有长度和类型限制

> 每一列的值或者单元格的值都具有时间戳,默认由系统制定,也可以由用户自己制定. 通过时间戳来区分不同版本的值, 一个单元格的不同版本按降序排列在一起

> 用户可以指定每个值所能保存的最大版本数


### 自动分区

- ** region加载数 **

> 最佳加载数10-1000, 每个region最佳大小为1GB-2GB
